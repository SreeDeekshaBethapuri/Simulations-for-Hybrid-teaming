In CIRL, Human is more of a teammate "who can play better" or has the true reward. 


The idea is to maximize expected reward via cooperation while inferring R. 

Model: Two agent POMDP, partially observable Mardov Decision process(decision making under uncertainity, not just about the future, but also the current state of the world.)


The agent doesnt know what state its in, instead, it maintains a belief over possible states.
