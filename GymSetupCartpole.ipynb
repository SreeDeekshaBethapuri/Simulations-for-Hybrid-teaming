```python
class Agent():
    def __init__(self, env):
        self.action_size = env.action_space.n

    def get_action(self, state):
        return 0 if state[2] < 0 else 1
```

Here, state[2] accesses the pole angle. If the pole is leaning to the left(pole angle < 0), the agent chooses action = 0-> move the cart left. 

if the pole is leaning towards right (pole angle is >= 0), it chooses action = 1 -> Move the cart right. 

```python
agent = Agent(env)
state, _ = env.reset()

for _ in range(200):
    action = agent.get_action(state)
    state, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    env.render()

env.close()
    # if done:
    #     break
```

```python
import gym

env = gym.make("CartPole-v1", render_mode="human")

class Agent():
    def __init__(self, env):
        self.action_size = env.action_space.n

    def get_action(self, state):
        return 0 if state[2] < 0 else 1

agent = Agent(env)
state, _ = env.reset()

for _ in range(200):
    action = agent.get_action(state)
    state, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated
    env.render()

env.close()
```
